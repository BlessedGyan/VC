```python


# Define file paths using raw string notation
cab_data_path = r"C:\Users\bless\Downloads\Cab_Data.csv"
customer_data_path = r"C:\Users\bless\Downloads\Customer_ID.csv"
transaction_data_path = r"C:\Users\bless\Downloads\Transaction_ID.csv"
city_data_path = r"C:\Users\bless\Downloads\City.csv"

# Load the datasets using the provided file paths
cab_data = pd.read_csv(cab_data_path)
customer_id = pd.read_csv(customer_data_path)
transaction_id = pd.read_csv(transaction_data_path)
city = pd.read_csv(city_data_path)

# Merging all datasets together based on foreign keys
master_data = (transaction_id.merge(cab_data, how='inner')).merge(customer_id, how='inner').merge(city, how='inner')

# Inserted new column of profits made with one transaction
master_data.insert(loc=9, column='Profit', value=master_data['Price Charged'] - master_data['Cost of Trip'])

# Inserted price of per KM of travel
master_data.insert(loc=6, column='Price per KM', value=master_data['Price Charged'] / master_data['KM Travelled'])

# Inserted price of per KM of travel
master_data.insert(loc=7, column='Profit per KM', value=master_data['Profit'] / master_data['KM Travelled'])

# Derive the year from the 'Date of Travel' column
index = 0
for date in master_data['Date of Travel']:
    days_since_2016 = date - 42364  # Assuming Date of Travel is in Excel format
    if days_since_2016 >= 0 and days_since_2016 <= 335:
        master_data.at[index, 'Year'] = 2016
    elif days_since_2016 <= 700:
        master_data.at[index, 'Year'] = 2017
    else:
        master_data.at[index, 'Year'] = 2018
    index = index + 1

# Count the number of rides per customer
rides_count = master_data['Customer ID'].value_counts().rename_axis('Customer ID').reset_index(name='Number of Rides')

# Merge the ride counts with the master data
master_data = pd.merge(master_data, rides_count, on='Customer ID')

# Display the final merged master data
master_data
```

```python
# Calculate CLV (Customer Lifetime Value)
customer_clv = master_data.groupby('Customer ID')['Profit'].sum().reset_index().rename(columns={'Profit': 'CLV'})
master_data = pd.merge(master_data, customer_clv, on='Customer ID')
```

```python
from sklearn.cluster import KMeans

# KMeans Clustering based on Rides and CLV
customer_segmentation = master_data[['Customer ID', 'Number of Rides', 'CLV']].drop_duplicates()
kmeans = KMeans(n_clusters=4, random_state=42)
customer_segmentation['Segment'] = kmeans.fit_predict(customer_segmentation[['Number of Rides', 'CLV']])
```

```python
### Null Hypothesis (H0): There is no significant difference in the average profit based on Payment Mode (Cash vs. Card) for both cab companies.
# Alternative(H1): There is a significant difference in the average profit based on Payment Mode for at least one of the cab companies.

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Filter the data for each company
pink_cab_data = master_data[master_data['Company'] == 'Pink Cab']
yellow_cab_data = master_data[master_data['Company'] == 'Yellow Cab']

# Step 2: Calculate the average profit for each Payment Mode for both companies
pink_cab_avg_profit = pink_cab_data.groupby('Payment_Mode')['Profit'].mean()
yellow_cab_avg_profit = yellow_cab_data.groupby('Payment_Mode')['Profit'].mean()

# Step 3: Independent t-test for Pink Cab
pink_cash_profit = pink_cab_data[pink_cab_data['Payment_Mode'] == 'Cash']['Profit']
pink_card_profit = pink_cab_data[pink_cab_data['Payment_Mode'] == 'Card']['Profit']

t_stat_pink, p_value_pink = stats.ttest_ind(pink_cash_profit, pink_card_profit, equal_var=False)

# Step 4: Independent t-test for Yellow Cab
yellow_cash_profit = yellow_cab_data[yellow_cab_data['Payment_Mode'] == 'Cash']['Profit']
yellow_card_profit = yellow_cab_data[yellow_cab_data['Payment_Mode'] == 'Card']['Profit']

t_stat_yellow, p_value_yellow = stats.ttest_ind(yellow_cash_profit, yellow_card_profit, equal_var=False)

# Step 5: Visualize the differences using a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Payment_Mode', y='Profit', hue='Company', data=master_data, palette='Set2')
plt.title('Profit Distribution by Payment Mode for Pink Cab and Yellow Cab')
plt.xlabel('Payment Mode')
plt.ylabel('Profit')
plt.grid(True)
plt.show()

# Step 6: Print the results of the t-tests
print("Pink Cab - T-statistic:", t_stat_pink, "P-value:", p_value_pink)
print("Yellow Cab - T-statistic:", t_stat_yellow, "P-value:", p_value_yellow)

### Similarly, the p-value (0.294) is greater than 0.05. We fail to reject the null hypothesis,
#   suggesting that there is no significant difference in the average profit between Cash and Card payments for Yellow Cab.
```
